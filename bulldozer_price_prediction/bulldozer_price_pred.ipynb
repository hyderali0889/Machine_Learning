```python
# Filling null values
# Numeric Values First

def convert_to_int(list): 
    # Converting strings to Catergories
    for label,feature in list.items():
        if(pd.api.types.is_string_dtype(feature.dropna())):
            list[label] = feature.astype("category").cat.as_ordered()
# Filling null numeric values
    
    for label,feature in list.items():
        if(pd.api.types.is_numeric_dtype(feature)):
            if(pd.isnull(feature).sum()):
                list[label] = feature.fillna(feature.median())  
# Filling null Categorical Values values
    
    for label,feature in list.items():
        if not pd.api.types.is_numeric_dtype(feature):
            list[label] = pd.Categorical(feature).codes + 1   
```

# Machine Learning Model Training

```python
df_tmp.info()
```

```python
# Spllitting data into train and valid split

training_set = df_tmp[df_tmp.year != 2012]
valid_set = df_tmp[df_tmp.year == 2012]
```

```python
x_train,y_train = training_set.drop("SalePrice" , axis =1 ) , training_set.SalePrice
x_valid,y_valid = valid_set.drop("SalePrice" , axis =1 ) , valid_set.SalePrice

x_train.shape , y_train.shape , x_valid.shape , y_valid.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_log_error , r2_score
def rmsle(y_true , y_preds): 
    return np.sqrt(mean_squared_log_error(y_true , y_preds))

def show_metrics(model):
    train_preds = model.predict(x_train)
    valid_preds = model.predict(x_valid)
    
    metrics = { "train rmsle" : rmsle(y_train , train_preds) , 
                "valid rmsle" : rmsle(y_valid , valid_preds) ,      
                "train mean Absolute Error" : mean_absolute_error(y_train , train_preds),
                "valid mean Absolute Error" : mean_absolute_error(y_valid , valid_preds),     
                "train mean Squared Log Error" : mean_squared_log_error(y_train , train_preds),
                "valid mean Squared Log Error" : mean_squared_log_error(y_valid , valid_preds),
                "train r2_score" : r2_score(y_train , train_preds),
                "valid r2_score" : r2_score(y_valid , valid_preds)
              }
    return metrics
    
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor( n_jobs=-1 , random_state=42 , max_samples=10000 , )
```

```python
%%time
model.fit(x_train , y_train)
```

```python
show_metrics(model)
```

```python
from sklearn.model_selection import RandomizedSearchCV

grid = { "max_depth" : [1000 , 800 , 500] , "max_features" : ["sqrt", "log2"] , "min_weight_fraction_leaf" : [0.1 , 0.3 , 0.4]  , "min_samples_split" :[15,10,8]}

rc_model = RandomizedSearchCV(  model , grid , n_iter=5 , cv=5 ,verbose=2)
```

```python
rc_model.fit(x_train , y_train)
```

```python
rc_model.best_params_
```

```python
show_metrics(rc_model)
```

```python
# Most Ideal Params

ideal_model = RandomForestRegressor( n_estimators=40 , min_samples_leaf=1 , min_samples_split=14 , max_features=0.5 , n_jobs=-1 , max_samples=None , random_state = 20 )
```

```python
ideal_model.fit(x_train,y_train)
```

```python
show_metrics(ideal_model)
```
