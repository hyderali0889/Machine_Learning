# Plot from dict

```python
store = { "lays" : 20 , "pepsi" : 12 , "cookies": 30 }
```

```python
fig,ax = plt.subplots()
ax.barh( list(store.keys() ), list(store.values()))
ax.set( ylabel="Products" , xlabel="Price" , title="My Store" );
```

# Plotting on multiple axis

```python
fig , (( a1, a2 ) , (a3,a4)) = plt.subplots( nrows=2 , ncols=2 , figsize=(10,10) )
np.random.seed(0)
q= np.random.randn(120)

a1.plot( q , q * 2 )
a2.scatter(q , q //2)
a3.bar(store.keys() , store.values() )

a4.hist(q);
```

# Plotting from a dataframe

```python
sales =pd.read_csv("car-sales.csv")
sales
```

```python
sales["Price"] = sales["Price"].str.replace('[\$\,]|\.\d*', '',regex=True).astype(int)
```

```python
sales
```

```python
sales["Date"] = pd.date_range( "1/1/2022" , periods=len(sales))
```

```python
sales
```

```python
sales["Total Sales"] = sales["Price"].cumsum()
```

```python
sales
```

```python
sales.plot(x="Date" , y= "Price")
```

```python
sales.plot( x="Total Sales" , y="Date" , kind="scatter" )
```

```python
sales.plot.hist(bins=50)
```

```python
sales.plot( x="Make" , y="Odometer (KM)" , kind="bar" ,)
```

```python

sales.plot(  y="Price" , kind="pie" )
```
