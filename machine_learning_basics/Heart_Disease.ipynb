```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
main= pd.read_csv("heart-disease.csv")
main.head()
```

```python
# Pyplot Method

main["age"].plot.hist(bins=50)


ab50 = main[main["age"] > 50]
```

```python
ab50
```

```python
# OO Method with pyplot



fig,ax = plt.subplots(figsize=(10,6))

ab50.plot.scatter( x="age" , y="trestbps" , c="target" , ax=ax )

ax.set_xlim(20 , 100)

```

```python
#OO Method

fig,(a0,a1 , a2) = plt.subplots( nrows=3, ncols=1 , figsize=(10,15)  , sharex=True)

scatter=a0.scatter(ab50['age'] , ab50["trestbps"] , c = ab50['target'])
a0.set( title="Heart Disease Prediction"  , ylabel="Trestbps"  );
a0.legend(*scatter.legend_elements() , title="Targets");
a0.axhline(ab50['trestbps'].mean() , linestyle="dashdot");

scatter1=a1.scatter(ab50['age'] , ab50["chol"] , c = ab50['oldpeak'])
a1.set( title="Heart Disease Prediction with cholesrtrol" , xlabel="Age" , ylabel="Chol"  );
a1.legend(*scatter1.legend_elements() , title="Old Peak");
a1.axhline(ab50['chol'].mean() , linestyle=":");

scatte2 = a2.hist(ab50["age"])
a2.set(title=" Age Historam" , xlabel="Age"  );

a2.set_ylim(0,50)

a2.axhline(ab50['age'].mean() , linestyle=":");

fig.suptitle("Heart Disease Analysis" , fontsize=20, fontweight="bold");

```
